**Handling outliers**

In the cells below, a function was derived to look for outliers within the weather features and tweak them into the data range. Outliers can be meaningless aberrations caused by measurement and recording errors. In such case, they can cause problems or lead to poor analysis of the effects of weather features on loadshort_fall, so it's important to question and analyze outliers.

```python
def sub_outliers_pressure(df, col):
    upper_limit = df[col].mean() + 3*df[col].std()
    lower_limit = df[col].mean() - 3*df[col].std()
    df[col] = np.where(df[col] > 1052,
        upper_limit,
        np.where(
            df[col] < 950,
            950,
            df[col]
        )
    )
    return df.describe()
```

```python
def sub_outliers_wind(df, col):
    upper_limit = df[col].mean() + 3*df[col].std()
    lower_limit = df[col].mean() - 3*df[col].std()
    df[col] = np.where(df[col] > 19,
        upper_limit,
            df[col]
    )
    
    return df.describe()
```

**Creating new features**

It was discovered during exploratory data analysis that the wind degree of barcelona and bilbao cities were stored in a categorical form while the wind degree of other cities are stored as integers. So there was a need to transform the categorical levels in barcelona and bilbao to integers because the model preferably handles numerical data. The cells below change the category into numerics.

```python
#creating a level category for the wind degree columns
classes = {
    "level_1" : [0,36],
    "level_2" : [36,72],
    "level_3" : [72,108],
    "level_4" : [108,144],
    "level_5" : [144,180],
    "level_6" : [180,216],
    "level_7" : [216,252],
    "level_8" : [252,288],
    "level_9" : [288,324],
    "level_10" : [324,360]
}
```

```python
#A function that changes the wind degree columns to categorical
def change_level(df, column_name, dictionary):
    row = 0
    for x in df[column_name]:
        for key, value in dictionary.items():
            if x == 0.0:
                df.at[row, column_name] = key
            if x > value[0] and x <= value[1]:
                df.at[row, column_name] = key
        row += 1
    
    return
```

```python
change_level(df_train, "barcelona_wind_deg", classes)
```

```python
change_level(df_train, "bilbao_wind_deg", classes)
```

**Checking highly correlated columns**

The function below identifies weather feature columns that are highly correlated, meaning they are some what linearly dependent with other features. A correlation matix was computed to identify such columns, followed by the introduction of lambda function to drop those columns that are highly correlated. It was neccessary to do all these because those features contribute very less in predicting the "loadshortfall_3h" but increses the computational cost.

```python
# Function for choosing highly correlated features to drop

def identify_correlated(df, threshold):
    """
    A function to identify highly correlated features.
    """
    # Compute correlation matrix with absolute values
    matrix = df.corr().abs()
    
    # Create a boolean mask
    mask = np.triu(np.ones_like(matrix, dtype=bool))
    
    # Subset the matrix
    reduced_matrix = matrix.mask(mask)
    
    # Find cols that meet the threshold
    columns_to_drop = [c for c in reduced_matrix.columns if \
              any(reduced_matrix[c] > threshold)]
    
    return columns_to_drop
```
