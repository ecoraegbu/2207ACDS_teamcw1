**Preparing the model**

```python
y = df_clean[:len(df_train)][['load_shortfall_3h']]
x = df_clean[:len(df_train)].drop('load_shortfall_3h', axis = 1)

x_train = df_clean[:len(df_train)].drop('load_shortfall_3h', axis = 1)
x_test = df_clean[:len(df_train)].drop('load_shortfall_3h', axis = 1)
```

**Loading the model**

```python
lr = LinearRegression()
```

```python
#normal linear model
#Using just the training dataset to test the model accuracy
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x_train
```

```python
lr.fit(x_train,y_train)

predictions = lr.predict(x_test)
```

**Evaluating the Model: Linear Regression Model**

```python
def rmse(y_test, y_predict):
    return np.sqrt(mean_squared_error(y_test, y_predict))
```

```python
rmse(y_test, predictions)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, predictions)
```
